```python
import numpy as np 
import pandas as pd

np.random.seed(42)

#have 2 groups for purchases
control_sales = np.random.normal(loc=45, scale=15, size=500)
test_sales = np.random.normal(loc=52, scale=18, size=500)

df_control = pd.DataFrame({'group':'control','purchase_amount':control_sales})
df_test = pd.DataFrame({'':'test', 'purchase_amount': test_sales})
df=pd.concat([df_control,df_test],ignore_index=True)

print(df.head(20))

#Have to replace the values which are located below 0 with 0 
df.loc[df['purchase_amount']<0,'purchase_amount']=0
df.sort_values(by='purchase_amount',ascending=True)




```

```python
##Stage2: Creating the brain of A/B
#Normality check
from scipy.stats import shapiro
control_data = df[df['group']=='control']['purchase_amount']
test_data= df[df['group']=='test']['purchase_amount']

#Shapiro Test
p_val_control= shapiro(control_data).pvalue
p_val_test = shapiro(test_data).pvalue

print(f"Shapiro-Wilk Test - Control Group P-Value: {p_val_control:.4f}")
print(f"Shapiro-Wilk Test - Test Group P-Value: {p_val_test:.4f}")

#**Distributions of the groups data are normal (p>alph), so ı'll go with parametric tests.

# Homogenity of variances chech (Levene Test)
#If the variances are equal, we use a Standard Independent T-Test.If not, the algorithm must adjust and use a Welch's T-Test instead.
from scipy.stats import levene
p_val_levene = levene(control_data,test_data).pvalue

print('p_val_levene: ', p_val_levene)

#** So, variances not equal(p<0.05), then ı use Welsch's T-test
#** If there was homogenity for our groups' variances , then ı would have to use Standard T-test
#** Null Hypothesis: There is no difference in the average purchase amounts between the old design and the new design.They are statistically equal.
#Welchs's T-Test 
from scipy.stats import ttest_ind
p_val_ttest = ttest_ind(control_data,test_data, equal_var=False).pvalue
print(f"Welchs's T-Test p value:{p_val_ttest:.4f}")

```
